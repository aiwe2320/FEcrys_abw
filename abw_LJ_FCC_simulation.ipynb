# LJ Solid FCC Simulations
An attempt to gather simulation data using FEcrys for an FCC LJ solid for use with the NN aspect of the code

```python
from O.MM.sc_system import *
```

```python
# Parameters
name = 'lj'
n_atoms_mol = 1
T = 60
P = 50 / 101.325  # kPa to atm
FF_name = 'OPLS'
atom_order_PDB_match_itp = False  # What do?
PME_cutoff = 0.8
form = 'FCC'
cell = '223'  # An approximation
```

```python
# Run equilibration:
PDB_ideal_supercell = f'./O/MM/GAFF_sc/lj/lj_fcc_unitcell_cell223.pdb'
sc = SingleComponent(PDB = PDB_ideal_supercell, n_atoms_mol=n_atoms_mol, name=name, FF_name=FF_name, atom_order_PDB_match_itp=atom_order_PDB_match_itp)
sc.initialise_system_(PME_cutoff=PME_cutoff, nonbondedMethod=app.LJPME)  # Note change to LJPME
sc.initialise_simulation_(timestep_ps=0.005, P=P, T=T, barostat_type=2)  # Pressure in atmosphere, using MonteCarloAnisotropicBarostat (like in cmp)
sc.run_simulation_(20000, 50)
sc._xyz = tidy_crystal_xyz_(sc.xyz, sc.boxes, n_atoms_mol=sc.n_atoms_mol, ind_rO=10)
sc.save_simulation_data_(f'./O/MM/GAFF_sc/{name}/data/{name}_{FF_name.lower()}_NPT_dataset_Form_{form}_Cell_{cell}_Temp_{T}')
index = get_index_average_box_automatic_(sc.boxes)
sc.save_pdb_(sc.xyz[index], b=sc.boxes[index], name=f'./O/MM/GAFF_sc/{name}/{name}_{FF_name.lower()}_equilibrated_Form_{form}_Cell_{cell}_Temp_{T}.pdb')
print('finished:',form,cell)
del sc
```
